```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
from plotly.offline import iplot
import plotly as py
```

```python
import cufflinks as cf
```

```python
py.offline.init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.randn(100,3),columns=['A','B','C'])
df.head()
```

```python
df = df.cumsum()
df.iplot()
```

```python
df.iplot(kind='scatter',x='A',y='C',mode='markers',title='Scatter Plot',xTitle='X',yTitle='Y')
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.iplot(kind='bar',x='sex',y='survived',title='sex Vs survived',xTitle='SEX',yTitle='SURVIVED')
```

```python
df.iplot(kind='bar',barmode='stack',bargap=0.5)
```

```python
df.iplot(kind='barh',barmode='stack',bargap=0.5)
```

```python
df.iplot(kind='box')
```

```python
df.iplot(kind='area',fill=True)
```

```python
df.iplot(kind='surface')
```

```python
cf.datagen.sinwave(10,0.25).iplot(kind='surface',colorscale='')
```

```python
df.iplot(kind='spread')
```

```python
df.iplot(kind='hist',bins=25,barmode='overlay')
```
