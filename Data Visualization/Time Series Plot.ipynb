```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:/Data Science/Data Visualization/DataVisualization/Time Series Plot/daily-temperature.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df.set_index('Date',inplace=True)
```

```python
df
```

```python
df.plot(style='.',color='red')
plt.ylabel("Temperature")
plt.title("Date Vs Temperature")
plt.legend(loc=1)
plt.show()
```

```python
groups = df.groupby(pd.Grouper(freq='A'))
```

```python
keys = groups.groups.keys()
```

```python
for key in keys:
    print(key)
```

```python
years = pd.DataFrame()
for key in keys:
    years[key] = groups.get_group(key)['Temp'].values
```

```python
years
```

```python
years.plot(subplots=True,figsize=(10,30))
plt.show()
plt.tight_layout()
```

```python
plt.matshow(years.T,aspect='auto')
plt.show()
```
