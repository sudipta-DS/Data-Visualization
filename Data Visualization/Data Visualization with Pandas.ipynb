```python
### Scatter_Matrix and Subplots

from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(df,figsize=(20,20),diagonal='kde',color='r')
plt.show()
```

```python
df.plot(subplots=True,sharex=False)
plt.tight_layout()
plt.show()
```

```python
df.plot(subplots=True,sharex=False,figsize=(6,6),layout=(2,2))
plt.tight_layout()
plt.show()
```
