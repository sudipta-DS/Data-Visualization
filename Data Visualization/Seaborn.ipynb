```python
def sinplot(flip=1):
    x = np.linspace(1,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*0.5)*(7-i)*flip)
```

```python
sinplot()
```

```python
sns.set_style('ticks',{'axes.grid' : True,'xticks.direction':'in'})
sinplot()
sns.despine(left=False,bottom=False,top=False,right=False)
```
