```python
## Nested Pie Plots

fig,ax = plt.subplots(dpi=100)
size = 0.3

values = np.array([[20,35],[56,70],[90,85]])
vals_sum = values.sum(axis=1)
vals_flat = values.flatten()

cmap = plt.get_cmap('tab20c')
outer_colors = cmap(np.arange(3)*4)
inner_colors = cmap(np.arange(6)*3)

ax.pie(vals_sum,radius=1,colors = outer_colors,wedgeprops=dict(width=size,edgecolor='w'))
ax.pie(vals_flat,radius=1-size,colors = inner_colors,wedgeprops=dict(width=size,edgecolor='w'))
plt.show()
```

### Polar Co-ordinates

```python
np.random.seed(0)
N = 20
theta = np.linspace(0.0,2*np.pi,N,endpoint=False)

radii = 10*np.random.rand(N)
width = (np.pi/4)*np.random.rand(N)
colors = plt.cm.plasma(radii/10)

plt.figure(dpi=150)
ax = plt.subplot(1,1,1,projection='polar')
ax.bar(theta,radii,bottom=0,width=width,color=colors,alpha=0.7)
plt.show()
```

```python
r = np.arange(0,2,0.01)
theta = 2*np.pi*r

plt.figure(dpi=100)
ax = plt.subplot(111,projection='polar')
ax.plot(theta,r)
ax.set_rmax(2)
ax.set_rticks([0,0.5,1,1.5])
ax.set_rlabel_position(-30)
plt.show()
```

```python
np.random.seed(2)
fig = plt.figure(dpi=100)
ax = fig.add_subplot(111,projection='polar')

r = 2*np.random.rand(150)
theta = 2*np.pi*np.random.rand(150)
area = 200*(r**2)
colors = theta

c =ax.scatter(theta,r,s=area,c=colors,cmap='hsv',alpha=0.7)
```

```python
from matplotlib.patches import Polygon
```

```python
def Fun(X):
    return (X-4)*(X-5)*(X-7)+85
a,b = 2,9
x = np.linspace(0,10)
y = Fun(x)

fig,ax = plt.subplots(dpi=150)
ax.plot(x,y,color='r',linewidth=2)
ax.set_ylim(bottom=0)

ix = np.linspace(a,b)
iy = Fun(ix)
verts = [(a,0),*zip(ix,iy),(b,0)]
poly = Polygon(verts,facecolor='0.9',edgecolor='0.5')
ax.add_patch(poly)
ax.text(0.5*(a+b),30,r"$\int_a^b f(x)dx$")
plt.show()
```
